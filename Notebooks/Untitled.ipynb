```python
import os

def get_full_extension(file_path):
    extensions = []
    while True:
        file_name, file_ext = os.path.splitext(file_path)
        if file_ext:
            extensions.append(file_ext)
            file_path = file_name
        else:
            break
    return ''.join(reversed(extensions))

file_path = '/home/sangu/stipskull.hdr'
file_ext = get_full_extension(file_path)

print(file_ext)  # Output: .nii.gz
```

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# Load MRI scan and mask
mri_path = "/home/sangu/Medulu/Datasets/Raw/post_reg_mri/coreg_IXI048-HH-1326-T1_ss.nii.gz"
mask_path_3 = "/home/sangu/rought_folder/trial/cr_ants_mask_3.nii.gz"
mask_path_1 = "/home/sangu/rought_folder/trial/cr_ants_mask_1.nii.gz"
mask_path_0 = "/home/sangu/rought_folder/trial/cr_ants_mask_0.nii.gz"

mri_img = nib.load(mri_path)
mask_img_3 = nib.load(mask_path_3)
mask_img_1 = nib.load(mask_path_1)
mask_img_0 = nib.load(mask_path_0)

mri_data = mri_img.get_fdata()
mask_data_3 = mask_img_3.get_fdata()
mask_data_1 = mask_img_1.get_fdata()
mask_data_0 = mask_img_0.get_fdata()

# Get mid-slice index
mid_slice = mri_data.shape[0] // 2

# Plot MRI with mask overlay
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(mri_data[mid_slice, :, :], cmap="gray")
plt.title("Original MRI")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(mri_data[mid_slice, :, :], cmap="gray")
plt.imshow(mask_data_3[mid_slice, :, :], cmap="plasma", alpha=0.4)  # Overlay mask in color
plt.imshow(mask_data_1[mid_slice, :, :], cmap="jet", alpha=0.4)  # Overlay mask in color
plt.imshow(mask_data_0[mid_slice, :, :], cmap="Blues", alpha=0.4)  # Overlay mask in color
plt.title("MRI with Mask Overlay")
plt.axis("off")

plt.show()
```
